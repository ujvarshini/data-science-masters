Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

A pandas Data frame is defined as a table which can have multiple columns and it is heterogeneous whereas Pandas Series is a 1D array of data with homogeneous dataset.

Example of a pandas dataframe is as below

```python
## Pandas dataframe

import pandas as pd
data2 = {'name':['ash','bob'],'age':["23","25"]}
df4 = pd.DataFrame(data2)


## pandas series
df5 = pd.Series(data2)
```

Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

DataFrames are both value and size-mutable (A Series, by contrast, is only value-mutable, not size-mutable. The length of a Series cannot be changed although the values can be changed).

Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
'''
three series are created

'''
# Create Series by assigning names
courses = pd.Series(["Spark","PySpark","Hadoop"], name='courses')
fees = pd.Series([22000,25000,23000], name='fees')
discount  = pd.Series([1000,2300,1000],name='discount')
```
