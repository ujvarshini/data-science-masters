Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

```python
'''
ArithmeticError is simply an error that occurs during numeric calculations.

ArithmeticError types in Python include:

OverFlowError
ZeroDivisionError
FloatingPointError
'''
### A. ZERODIVISION ERROR
try:
    1/0
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

```python
### OVERFLOW ERROR
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
'''
The LookupError exception in Python forms the base class for all exceptions that are raised when an 
index or a key is not found for a sequence or dictionary respectively.

'''

### index error An IndexError is raised when a sequence reference is out of range.

# lists
x = [1, 2, 3, 4]
try:
    print(x[10])
except LookupError as e:
    print(f"{e}, {e.__class__}")
```

```python
###key error

pylenin_info = {'name': 'Lenin Mishra',
                'age': 28,
                'language': 'Python'}
user_input = input('What do you want to learn about Pylenin==> ')

try:
    print(f'{user_input} is {pylenin_info[user_input]}')
except LookupError as e:
    print(f'{e}, {e.__class__}')
```

Q6. List down some best practices for exception handling in python.

Use try-except-finally blocks to handle exceptions gracefully and ensure your code can handle errors effectively. Catch specific exceptions rather than using a catch-all except block to avoid swallowing unexpected exceptions and make it easier to identify and fix issues

Consider raising custom exceptions when appropriate to provide more informative error messages to users.
Avoid common pitfalls, such as using overly broad exception handling or ignoring important exceptions, to ensure the code remains robust and reliable.
Embrace logging as a powerful tool for monitoring and debugging your applications, and learn to log exceptions and errors effectively to aid in troubleshooting.
